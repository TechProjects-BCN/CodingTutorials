Arduino and python can interact via Serial, a communication method you can check out right here if interested: https://en.wikipedia.org/wiki/Serial_communication

To start with, let's look at the code needed in our arduino:
```
void setup()
{
    Serial.begin(115200); // Serial Communication is starting with 115200 of baudrate speed
}

void loop()
{
    data = String(1) + "," + String(2) + "," + String(3) + "," + String(4); // in our case, we are just seding numbers
    Serial.println(data);
    delay(1000); // 1s delay
}
```

This code starts the serial communication and prints data into the serial port every 1 second. How can we retrieve this information?

````python
import serial

PORT = "COM1"
BAUDRATE = 115200

arduino = serial.Serial(port=PORT, baudrate=BAUDRATE, timeout=.1) # Initiate Communication from the computer
````

To help, we'll define some useful functions to retrieve our data and send new:

````python
def arduino_read(arduino):
        """
        Reads data from arduino
        :return: String
        """
        arduino.write(bytes("A", 'utf-8'))
        #sleep(0.7)
        return arduino.readline().decode('utf-8')
````

````python
import time

def arduino_write(arduino, x):
        arduino.write(bytes(x, 'utf-8'))
        time.sleep(0.05)
        return arduino.readline()
````
