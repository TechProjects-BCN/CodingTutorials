# Datatypes

A Datatype is the type of the data we can store in variables. In all programming languages, we call variable any characters that represent a value.
For example if we say:
```
x = 5
```
we are storing the value 5 inside x, so from now on, x will be 5.

There are multiple types of variables that you'll see below. 

Lists are multiple values stored together in one variable (e.g. [0, 1, 2, 3]). You can access each value by using ```[]``` after the variable that stores the list. 

Dictionaries are like lists, but values are stored with a key. So all values need a key, that is just a string, that will represent the value.

For example:
```
x = {"One": 1, "Two": 2}
```
if I do:
```
x["One"] # This will return to me the number 1
```
To store new values inside the dictionary, you can do as follows:
```
x["Three"] = 3
```
Now x is = {"One": 1, "Two": 2, "Three":3}

A set is like a list but can only contain unique values

````python
hw = "Hello World" # String
a = 1 # Integer
b = 2.54 # Float
c = 'c' # Character

l = [1, "Hello", 1.23] # List
d = {"John": 5, "Maria": 0.2, "Philip": "Other"} # Dictionary
t = (1, 2.0, "Hydrogen") # Tuple
s = {"Barcelona", "Valencia", "Madrid"} # Set

# To access data inside a variable like a list, dictionary, tuple or set, it has to be done as follows:

l[0] # == 1. will give you the first element. In python lists, sets and tuples are 0 indexed (first value is the 0th value)
l[1] # == "Hello". will return the second one

# to add an element to a dictionary, you can use:
l.append(4) # This will convert [1, "Hello", 1.23] to [1, "Hello", 1.23, 4]
# to remove an element, you can do it by id (which one is at a certain position), you can remove the last one, or you can remove a specific value
l.pop(0) # now l = ["Hello", 1.23, 4]
l.pop() # now l = ["Hello", 1.23]
l.remove(1.23) # now l = ["Hello"] instead of ["Hello", 1.23]

# Dictionaries work with keys and values. It works like this:
# {<Key>: <Values>}
# To Access the a value, you need to use the key:
d["John"] # == 5
d["Maria"] # == 0.2

# If you want a list containing all the keys or all the values, you can use
d.keys() # = ['John', 'Maria', 'Philip']
d.values() # = [5, 0.2, 'Other']

````

# Useful Functions

Functions are pieces of code that can be run by typing a name and a parenthesis after it. This will execute it and produce a result we may need.
A function can be thought of like the following:
Inputs (What's inside the parenthesis) -> Black Box (Something will be done with what you provided) -> Output (It may be a value that is returned or an action that is done)

````python
print("Hello World") # prints whatever is inside of it to the terminal
# if you want to print something with a variable inside:
name = "john"
print(f"Hello {name}")

# If you have a list
n = [0, 1, 2, 3, 4]
max(n) # gives the highest value in the list
min(n) # gives the smallest value in the list

# To create a list that counts from 0 to n-1
n = 5
range(n) # == [0, 1, 2, 3, 4]
# For setting starting number:
range(0, n) # == [0, 1, 2, 3, 4]
# If you want every other number
range(0, n, 2) # == [0, 2, 4]

# all of these functions return values, so you can store it inside a variable as follows:
ZeroToFour = range(n)
````

# Writting our custom Functions

````python
# The syntax to write our own function is as follows:

# def "name"(inputs):
#     <Code that does something> # Optional
#     return a, b, c, d, .... # We return the values we want

# An example would be a function that returns the value of the square of the number given
# aka. y = x^2

def y(x): # we set the name of the function to y, and say it will accept 1 input (x)
  return x**2 # We return the result of x^2

y(2) # == 4
y(4) # == 16

# We can write more complex functions, for example, to calculate the preassure at an altitude given Atmospheric Preassure at sea level, R (Gases Constant),...
def preassure(z, Po=101325, go=9.81, lmbda=-0.0065, To = 288.16, R=287, Zo=0): # here we are already defining values to the inputs. What this means, is that they are not mandatory to specify.
# So if the user doesn't say otherwise, the values will be these
  T = 1 + (lmbda*(z-Zo))/To
  exponent = -go/(lmbda * R)
  return Po * (T ** exponent)

preassure(5000) # preassure at 5km of altitude

# If we are in the Mesosphere (between 51km and 71km), the lambda, preassure, and To is different than in the troposphere (the default values), so we can specify that:
preassure(60000, lmbda=-0.0028, Zo=51000, To=270.65, Po=79.78)
````

# For Loops and while Loops

If we want to iterate through the same line several times, like if we want to print something twice or run some code until something happens, we need For and While loops. Here's an explanation of how they work:

````python
# If we want to run code 3 times:
for i in range(3):
    print(i)
# Notice that we can keep track of the number of iteration using i
````

````python
# If we want to iterate over a list
x = ["Hi", 0, 0.21]
for y in x:
    print(f"Value: {y}")
````

````python
# If we want to iterate over a list and keep track of the iteration
x = ["Hi", 0, 0.21]
for i, y, in enumerate(x):
    print(f"i: {i}; y: {y}")
````

What if we want to execute code indefinitely or until a condition is met? That's why While loops exist:

````python
# Run code indefinitely
while True:
    print("Hello world")
    break # This exits any loop (Both For and While loops). Without the break, this code would never stop 
````

````python
x = 0
while x < 10:
    print(x)
    x += 1 # increment x
````

# Graphing

If you want to plot a function or a list of values, there's a library to do so.

````python
# Installation : pip install matplotlib

import matplotlib.pyplot as plt # Here We are importing pyplot, a library inside of the matplotlib library (Yeah, that exists), and saying that we want to call it plt. So from now on, instead of typing matplotlib.pytplot
# We can just type plt.something

# If for example we want to graph the function x**2:
# we define x to be the range we want to represent
x = numpy.arange(start=-1000, stop=1000, step=1)
# we define our function. 
f = lambda x: x**2 
# that is the same as
def f(x):
    return x**2
# get y from x
y = [f(_x) for _x in x]
# that is the same as
y = []
for _x in x:
    y.append(f(_x))

# We plot our function
plt.plot(x, y) # Line 
# plt.scatter(x, y) # with dots
# plt.hist(x) # Histogram
# You can check all available plots at: https://matplotlib.org/3.5.3/gallery/index.html
````
